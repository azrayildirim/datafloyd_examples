# Data Visualization by Using Python "matplotlib" Library  

Visualization is the best way to understand the data we have. We can perform powerful and insightful visualizations on our data by using Python `matplotlib` library 

In this tutorial we will start by importing `matplotlib` library and proceed with a progressive approach. To put it in other way, we will start drawing the graphics in its simplest form and gradually improve them to make them better looking and much more informative.


```python
# importing the library
import matplotlib.pyplot as plt

fig_size = [16,9]

# Let's add the numerical numpy library too
import numpy as np
```

--------------

<div class="alert alert-info">
<b>LINE GRAPHS</b> 
</div>

<div class="alert alert-success">
Line graphs are the simplest type of graph. They serve to visualize the continous connection between two variables. Line graphs, as the name suggests, creates a line between successive data. Let's use a simple function to iluustrate this.
</div>

In its simplest form, let's create a "sinus" data and plot it. In the following steps, we make changes on this graph and develop the graph gradually.

```python
# From 0 to 2*pi
x = np.linspace(0,np.pi*2,50)
y = np.sin(x)

plt.figure(figsize=fig_size)
plt.plot(x,y)

plt.show();
```

>Adding axes labels, title and grid

As you can see, even this simple graphic tells a lot, but let's improve this graph step by step:
* Add axis names and title to the chart.
* Add grid
Please note that these are not specific to the line graphs, the same functions can be used for other chart types.

```python
plt.figure(figsize=fig_size)
plt.plot(x,y)

# adding the ylabel
plt.ylabel("variable y")
# adding the x label
plt.xlabel("variable x")
# Adding the title
plt.title("Sample line");
# Adding the grid
plt.grid()

plt.show();
```

>Adding markers, legend and a second line

* Let's add markers to the data points
* Let's add a second line with red color and set the thickness of the line to 5 pixels
* Add a legend

```python
plt.figure(figsize=fig_size)

# adding markers to data points
plt.plot(x,y,'-ob')
# adding a second line with a dashed style
plt.plot(x,y*0.3,'--r',linewidth=5)


# adding y label
plt.ylabel("variable y")
# adding x label
plt.xlabel("varible x")
# adding title
plt.title("Sample Line")
# adding grid
plt.grid()
# adding legend
plt.legend(["First line","Second line"])

plt.show();
```

>Custom ticks and labels for x axis, adding an annotation 

* Set the ticks of the x axis as 0, pi/2, pi, 3pi/2, 2pi and show them on the figure
* Let's put an annotation on the value of x=pi and y=0.

```python
plt.figure(figsize=fig_size)

# adding markers to data points
plt.plot(x,y,'-ob')
# adding a second line with a dashed style
plt.plot(x,y*0.3,'--r',linewidth=5)


# adding y label
plt.ylabel("variable y")
# adding x label
plt.xlabel("varible x")
# adding title
plt.title("Sample Line")
# adding grid
plt.grid()
# adding legend
plt.legend(["First line","Second line"])

# Using LaTeX for tick labels 
plt.xticks([0,np.pi/2,np.pi,1.5*np.pi,2*np.pi],[r"0",r"$\dfrac{\pi}{2}$","$\pi$",r"$\dfrac{3\pi}{2}$","$2\pi$"])

plt.annotate("Look at this!\n ($\pi$,0)",[np.pi,0],[1.5*np.pi,0.6],
             arrowprops=dict(facecolor='m', shrink=0.05,),fontsize=13,color="g")

plt.show();
```

After this basic introduction to line graphs let's continue to the next graph type

--------------------

<div class="alert alert-info">
<b>SCATTER GRAPH</b> 
</div>

<div class="alert alert-success">
These graphs are generally used to examine the relationship between the two variables. Let's use the infamous iris dataset to illustrate this graph type.
</div>

```python
# Loading iris dataset
from sklearn.datasets import load_iris

iris = load_iris()

# Making a new figure
plt.figure(figsize=fig_size)
# Let's use the first and the second feature
plt.scatter(iris.data[:,0],iris.data[:,1])
# Add labels for x and y axis
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.show();
```

>Adding color to the points and changing the size of points

* Let's color the points according to the classes in the iris data set.
* In addition, let's make the size of the points change according to the 3rd attribute

```python
plt.figure(figsize=fig_size)
# Let's use the first and the second feature
plt.scatter(iris.data[:,0],iris.data[:,1],c=iris.target,s=iris.data[:,2]*20)
# Add labels for x and y axis
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.show();
```

>Assigning colors to classes

In the figure above, it is not clear which color belongs to which class. For this we need to add a "legend". First, let's see how many classes are in the data set:

```python
iris.target_names
```

As you can see, there are 3 classes. Let's color setosa class as red, versicolor class as blue and virginica class as green . We'd better use a `for` loop for this. Thus we can assign the "legend" according to colors.

```python
# Colors for classes
colors = ["r","b","g"]

plt.figure(figsize=fig_size)

# Add labels for x and y axis
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

# drawing scatter plots for each class
for i,name in enumerate(iris.target_names):
    plt.scatter(iris.data[iris.target==i,0],iris.data[iris.target==i,1],s=iris.data[iris.target==i,2]*50,c=colors[i])
plt.legend(iris.target_names)

plt.show()
```

The chart isn't bad, but I hear you say, "What do we do if we want the sizes of the points in the legend to be homogenous?" Let's play a little with more code. And as a final bonus, let's chamge the shapes with some LaTeX code. For this we will use the standard equation format.

```python

colors = ["r","b","g"]
# instead of shapes we will have special charactes
shapes = [r'$\alpha$',r'$\beta$',r'$\gamma$']

plt.figure(figsize=fig_size)


# labelling x and y
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

# draw scatter for each class
for i,name in enumerate(iris.target_names):
    plt.scatter(iris.data[iris.target==i,0],iris.data[iris.target==i,1],s=iris.data[iris.target==i,2]*70,
                c=colors[i],
                marker=shapes[i],
               alpha=0.3)
lgnd = plt.legend(iris.target_names)

# Now let's fix the size for the legend
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
```

I think you understand the basic idea for the scatter graph. Now we can move on to another chart type.

<div class="alert alert-info">
<b>BAR PLOT</b> 
</div>

<div class="alert alert-success">
Bar plots are used mainly for comparing categorical data. Let's experiment and see how we can use these chart types.
</div>

Let's create some simple data to play with.

```python
plt.figure(figsize=fig_size)
y = [10,5,3,12,20]
x_labels = ['January','February','March','April','May']
x_pos = np.arange(len(y))

plt.bar(x_pos,y)

plt.title("Sales")

plt.xticks(x_pos,x_labels);

plt.xlabel("Months")
plt.ylabel("Sales amount (x1000)")
plt.show();
```

>Plotting the graph horizontally

We can plot the same graph horizontally by using `barh` instead of `bar` function

```python
plt.figure(figsize=fig_size)
y = [10,5,3,12,20]
x_labels = ['January','February','March','April','May']
x_pos = np.arange(len(y))

plt.barh(x_pos,y)

plt.title("Sales")

plt.xticks(x_pos,x_labels);

plt.xlabel("Months")
plt.ylabel("Sales amount (x1000)")
plt.show();
```

>Changing various properties of the bars

Let's make the following modifications to the figure:
* Assign each column a different color
* Assign different colors to the edges of the bars and make them thick
* Make the bars have different width sizes

```python
plt.figure(figsize=fig_size)
y = [10,5,3,12,20]
x_labels = ['January','February','March','April','May']
x_pos = np.arange(len(y))

plt.bar(x_pos,y,color=['black', 'red', 'green', 'blue', 'cyan'],
        edgecolor=["blue","green","cyan","black","red"],
       linewidth = 10,
       width = [0.5,0.6,0.7,0.3,0.4])

plt.title("Sales")

plt.xticks(x_pos,x_labels);

plt.xlabel("Months")
plt.ylabel("Sales amount (x1000)")
plt.show();
```

<div class="alert alert-info">
<b>STACKED/GROUPED BAR PLOT</b> 
</div>

<div class="alert alert-success">
The bars can be stacked or groupped in order to represent different needs. For examples you can see the total sales of various entities stacked on top of each other.<br>
Let's look at the groupped bar plot first</div>

```python
plt.figure(figsize=fig_size)
y_1 = [10,5,3,12,20]
y_2 = [5,8,9,10,11]
x_labels = ['January','February','March','April','May']
x_pos = np.arange(len(y))

plt.bar(x_pos-0.2,y_1,color="r",width=0.3)
plt.bar(x_pos+0.2,y_2,width=0.3)


plt.title("Sales")

plt.xticks(x_pos,x_labels);

plt.xlabel("Months")
plt.ylabel("Sales amount (x1000)")

plt.legend(["Product A","Product B"])
plt.show();
```

If we want to draw a stack of columns, we can give the "bottom" parameter as the series at the bottom.

```python
plt.figure(figsize=fig_size)
y_1 = [10,5,3,12,20]
y_2 = [5,8,9,10,11]
x_labels = ['January','February','March','April','May']
x_pos = np.arange(len(y))

plt.bar(x_pos,y_1,color="r")
plt.bar(x_pos,y_2,color="b",bottom=y_1)


plt.title("Sales")

plt.xticks(x_pos,x_labels);

plt.xlabel("Months")
plt.ylabel("Sales amount (x1000)")

plt.legend(["product A","product B"])
plt.show();
```

<div class="alert alert-info">
<b>HISTOGRAM </b> 
</div>

<div class="alert alert-success">
The histogram, a special kind of bar graph, is used to show the distribution of a variable.
</div>

Let's create a normally distributed variable

```python
number_of_samples = 10000

np.random.seed(42)

y = np.random.randn(number_of_samples)

plt.figure(figsize=fig_size)

plt.hist(y,bins=100);
```

<div class="alert alert-info">
<b> BOXPLOT </b> 
</div>

<div class="alert alert-success">
A "boxplot" graph gives important information about the distribution of the variable. "Median", "25th quartile", "50th quartile" values of the distribution can be seen on the graph.
</div>

Let's try to see

```python
number_of_samples = 10000

np.random.seed(42)

y = np.random.randn(number_of_samples)

plt.figure(figsize=fig_size)

ret = plt.boxplot(y);

#---------------------------
# Annotations
median = [item.get_ydata()[0] for item in ret['medians']]
plt.annotate("Median",[0.9,median[0]],[0.6,median[0]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")

boxes = [item.get_ydata() for item in ret['boxes']]
plt.annotate("Q1:25th quartile",[0.9,boxes[0][0]],[0.6,boxes[0][0]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")
plt.annotate("Q1:25th quartile",[0.9,boxes[0][0]],[0.6,boxes[0][0]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")
plt.annotate("Q3:75th quartile",[0.9,boxes[0][2]],[0.6,boxes[0][2]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")

caps = [item.get_ydata() for item in ret['caps']]
plt.annotate("Q3-1.5xIQR",[0.95,caps[0][0]],[0.6,caps[0][0]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")
plt.annotate("Q1-1.5xIQR",[0.95,caps[1][0]],[0.6,caps[1][0]],va='center',
         arrowprops=dict(facecolor='b'),fontsize=20,color="k")
#---------------------------

plt.title("Boxplot example")
plt.xticks([],[])
plt.show();
```
