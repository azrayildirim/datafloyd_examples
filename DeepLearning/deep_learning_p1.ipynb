Not: *numpy* ve *matplotlib* ile ilgili daha önceden hazırladığım yazıların bağlantılarını da vereyim ihtiyacınız olursa bakarsınız:<br>

```python
# Veri setini yükleyelim
fashion_mnist = keras.datasets.fashion_mnist
```

Şimdi yukarıdaki satırda bir nesne hazırlamış olduk, içindeki veriyi alabilmek için *load_data* metodunu kullanmamız gerekiyor. Eğer daha önceden veriyi indirmediyseniz bu metot bir indirme işlemi başlatacak.

```python
# Bu fonksiyon veriyi test ve train olmak üzere iki set halinde veriyor
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

Öncelikle verinin boyutlarına bakalım

```python
print("Eğitim setinin boyutu: ")
print(train_images.shape)
print("Test setinin boyutu: ")
print(test_images.shape)
```

Gördüğümüz gibi 60000 adet örnek eğitim setinde, 10000 adet örnek test setinde mevcut. Her bir imgenin boyutunun da 28x28 olduğunu görüyoruz. Sadece iki boyut olması imgelerimizin siyah-beyaz olduğu anlamına geliyor.

```python
# Eğitim ve test setindeki eşsiz sınıf sayısı
unique_classes,u_counts = np.unique(np.concatenate([train_labels,test_labels]),return_counts=True)
print(unique_classes)
print(u_counts)
```

Toplamda 10 adet sınıf var, ne yazık ki bu sınıfların hangi giysilere ait olduğu veri setinin içinde yazmıyor. Bunun için sınıfların yazdığı bir liste oluşturalım.

```python
class_names = ['Tişört / Üst', 'Pantolon', 'Kazak', 'Elbise', 'Ceket',
               'Sandal', 'Gömlek', 'Spor Ayakkabı', 'Çanta', 'Çizme']
```

Her sınıftan birkaç örnek görelim isterseniz. 

```python
num_of_samples_per_class = 10
num_classes = len(unique_classes)


fig, axes = plt.subplots(nrows=num_classes, 
                         ncols=num_of_samples_per_class, 
                         figsize=(num_classes,num_of_samples_per_class))

# Tüm sınıflar üzerinden dolaşalım
for ix,current_class in enumerate(unique_classes):
    
    # Örnek sayısı kadar dolaşalım
    for iy,current_sample in enumerate(range(num_of_samples_per_class)):
    
        # sıradaki örneğin indeksini bulalım
        index = np.where([train_labels == current_class])[1][current_sample]
        
        # imgeyi hazirlayalim
        current_image = train_images[index]
        
        # num_classes x num_of_samples kadar bir subplot yapalım
        axes[ix][iy].imshow(current_image, cmap=plt.cm.binary)
        axes[ix][iy].set_xticks([])
        axes[ix][iy].set_yticks([])
        

# Adding class names
for ax, row in zip(axes[:,0], class_names):
    ax.set_title(row, rotation=0, size='large')

plt.subplots_adjust(hspace=0.5,wspace=0.2)
```

Çok güzel hazırlanmış bir veri seti değil mi?Toplamda 10 sınıf ve her sınıftan 7000 adet örnek var. Kolay bir iş değil yapanlara selam ve saygılarımı gönderiyorum.

## Sınıflandırma yapabilmek için mimarinin hazırlanması

Her yapay sinir ağı çeşitli katmanlardan oluşur. Bu katmanlar temel olarak "girdi katmanı", "gizli katman(lar)" ve "çıktı katmanı" olarak gruplandırılabilir. Yukarıdaki veri setinde bulunan 10 sınıfı birbirinden ayırt edebilmek için işe basit bir mimari kurarak başlayalım.

*Not:Temel yapay sinir ağları bilgilerinizi tazelemek aşağıdaki yazılara bakmanızı tavsiye ederim:*<br>
<span style="color: #ff0000;"><strong><a style="color: #ff0000;" href="https://www.datafloyd.com/tr/yapay-sinir-aglari-neural-networks-1-bolum-tek-bir-noron-hakkinda/">Yapay Sinir Ağları 1.Bölüm</a></strong></span><br>
<span style="color: #ff0000;"><strong><a style="color: #ff0000;" href="https://www.datafloyd.com/tr/yapay-sinir-aglari-neural-networks-2-bolum/">Yapay Sinir Ağları 2.Bölüm</a></strong></span>

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

Yukarıdaki kodu inceleyelim:
* `keras.layers.Flatten(input_shape=(28, 28))` Bu katman tüm imgeyi tek boyutlu bir vektör haline dönüştürmektedir. Yani 2 boyutlu olan imge matrisimiz  $28\times28 = 784$ boyutlu bir vektöre dönüştürülmektedir. Bu işleme *"flatten"* yani düzleştirme denilmektedir. 
* `keras.layers.Dense(128, activation=tf.nn.relu)` Burada 128 adet adet düğümü olan bir katman yaratılıyor. *"Dense"* ifadesi katmandaki herbir düğümün bir önceki katmanın çıktıları ile birebir bağlantılı olduğunu göstermektedir. Yani hesaplayacak olursak bir önceki katmanda 784 adet düğüm mevcut. Her bir düğüm bir sonraki 128 katman ile bağlantı yapacak. Yani $784\times128=100352$ adet bağlantı mevcut. Her bir bağlantının aslında bir ağırlık olduğunu ve bunların eğitim sırasında güncellendiğini hatırlamakta fayda var. Unutmadan bu katmanın çıktısının *"Relu"* fonksiyonunundan geçtiğine de dikkat edin.
* `keras.layers.Dense(10, activation=tf.nn.softmax)` Bu katmanımız 10 adet düğüme sahip ve en soununda *softmax* fonksiyonu kullanılıyor. Bu fonksiyonunun ayrıntılarını ve matematiksel temellerini başka bir yazıda ele alacağım. Şimdilik bunun sınıf olasılıklarını oluşturmada kullanıldığını bilmemiz yeterli. Yani bu son katmanın çıktılarını topladığımız zaman 1 sayısına ulaşacağız. Buradaki bağlantı sayısını da hesaplayacak olursak sonuç  $100352\times10=1003520$ olmaktadır.

Gördüğünüz gibi bağlantı sayısı bu basit ağ yapısı için bile hemen 1 milyonun üzerine çıktı. Çok daha kompleks ağlarda toplam bağlantı sayısı 140-150 milyona kadar çıkabilmektedir. Sıra şimdi modelimizi derlemeye (compile) geldi

```python
model.compile(optimizer=tf.train.AdamOptimizer(), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

Burada da temel olarak 3 parametre görmekteyiz. Bunlar 
* `optimizer` Ağımızın parametrelerinin güncellenmesi için kullanılır. Bunu "loss" fonksiyonuna ve veriye bakarak yapar.  
* `loss` Optimizasyon sırasında modelin ne kadar doğru olduğunu ölçmede kullanılır.Eğitim sırasında bu metriği minimize etmeye çalışırız.
* `metrics` Eğitim ve test adımlarında tüm veri seti üzerinden çıkarılan bir metriktir. "Accuracy" örneklerin ne kadar doğru sınıflandırıldığını ölçer.

Modeli hazırlayıp derlediğimize göre şimdi geriye bu modeli eğitmek kaldı. Bu da *Keras* sayesinde oldukça basit bir işlem. Hızlı bir sonuç alabilmek için toplam 5 "epoch" seçelim. 1 "epoch" tüm eğitim setinin ileri ve geri yönde modelden 1 kere geçirilmesi olarak tanımlanmaktadır. 

## Normalizasyonun önemi

Sınıflandırmaya geçmeden önce çok önemli olan normalizasyon konusunu hatırlatmak istiyorum. Genel olarak hemen hemen tüm makine öğrenmesi problemlerinde girdi sağladığınız vektörleri normalize etmeniz gerekir. Bunu yapmanın çeşitli yöntemleri vardır. 

İmgeler için genel olarak tercih edilen yöntem  imgeyi alabileceği maksimum değere yani siyah-beyaz imgeler için 255'e bölerek normalize etmektir. 

Aman dikkat edelim eğitim ve test setinin normalizasyonunu aynı şekilde yapmamız gerekiyor yoksa bu "accuracy" neden artmıyor diye bakınıp dururuz :)

Normalizasyonun önemini daha iyi anlatabilmek adına <span class="burk">ilk önce imgeleri normalize etmeden modelimizi eğitelim</span>

```python
model.fit(train_images, train_labels, epochs=5)
```

Gördüğünüz gibi "accuracy" değerleri tabiri caizse yerlerde sürünüyor. Şimdi <span class="girk">normalize edelim</span> ve bakalım "accuracy" değerleri ne olacak?

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
model.fit(train_images, train_labels, epochs=5)
```

"Accuracy" değerlerine bakarsanız iki durum arasındaki farkın çok net olduğunu göreceksiniz.

Yani kıssadan hisse <span class="mark">sakın ama sakın verilerinizi normalize etmeyi unutmayın</span>.

## Modelin değerlendirilmesi

Keras'ın yazarı *François Chollet* Bey bizim için herşeyi düşünmüş, Keras ile hazırlanan modeli değerlendirmek o kadar kolay ki tek satır kod ile bunu gerçekleştirebilirsiniz. İnanmadıysanız yazıyorum: 

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
```

Değerlendirme bitti bu kadar :) Yukarıda tek satır sözü verdiğim için "print" kısmını yazmayarak bir mini çakallık gerçekleştirmiş olabilirim, onu da aşağıda yazıyorum.

```python
print("Test seti \"accuracy\" değeri {:.2f}".format(test_acc))
```

Vay be test bu kadar basit bir model ile bile göreceli olarak yüksek bir değer yakaladık. Bir sonraki adımda isterseniz tahminleri biraz daha yakından inceleyelim

## Tahminlerin incelenmesi

Öncelikle tüm test seti üzerinde tahminleme işlemini gerçekleştirelim.

```python
predictions = model.predict(test_images)
predictions.shape
```

Tahminler gördüğünüz gibi $10000\times10$ luk bir vektör olarak karşımıza çıktı. Yani tahmin edilen örnek sayısı çarpı sınıf sayısı kadar bir matrisimiz var. Bu matrisin her bir satırı o satırdaki örneğin hangi sınıfa ne kadar benzediğini ifade eden bir vektör. Bu vektör "softmax" çıktısı olduğu için toplamının bir olması gerekiyor. İsterseniz ilk satırda deneyerek öncelikle bunu doğrulayalım

```python
sum(predictions[0])
```

Tam bir değil ama neredeyse 1, o kadar kusur kadı kızında da olur diyelim. Ama bir de tüm elemanları toplamadan da geçmeyelim bu kısmı. Bunun da teoride 10000 etmesi gerekiyor. 

```python
sum(sum(predictions))
```

"Machine Precision" hatalarının kaçınılmaz olduğunu bir kenara not edelim ve bu tahminleri biraz görselleştirerek yolumuza devam edelim.

```python
# Tahminleri görselleştirmek için bir fonksiyon yazalım
def plot_predictions_and_image(sample_num):
    plt.figure(figsize=(7,3))
    plt.subplot(1,2,2)
    # Bar grafik güzel görünür
    bars = plt.barh(class_names,predictions[sample_num], color="#777777")
    
    # Tahmin edilen sınıfı, gerçek sınıfı ve güvenirliği bulalım
    pred_class = class_names[np.argmax(predictions[sample_num])]
    real_class = class_names[test_labels[sample_num]]
    confidence = np.max(predictions[sample_num])
    
    # Eğer doğru sınıflandırma gerçekleşirse mavi 
    # gerçekleşmezse kırmızı sütun çizelim
    if pred_class==real_class:
        bars[np.argmax(predictions[sample_num])].set_color("b")
    else:
        bars[np.argmax(predictions[sample_num])].set_color("r")
            
      
    plt.subplot(1,2,1)
    # imgeyi gösterelim
    plt.imshow(test_images[sample_num],cmap=plt.cm.binary)
    plt.xticks([])
    plt.yticks([])
    plt.title("Gerçekte: {:}".format(real_class) 
              + "\n" +
             "Tahmin edilen: {:} ({:.2f})".format(pred_class,confidence) )
    plt.subplots_adjust(wspace=0.5)
```
