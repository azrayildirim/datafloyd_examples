$W$, ve $x$ adında iki tane sütun matrisimiz ve bir adet $b$ adında sabit terimimiz olsun.
</div>

\begin{equation*}
W=
\begin{bmatrix}
w_0 \\
\vdots \\
w_n
\end{bmatrix}
\quad
x=
\begin{bmatrix}
x_0 \\
\vdots \\
x_n
\end{bmatrix}
\quad
b
\end{equation*}

Şimdi $y=W^Tx+b$ yazdığımızda ne olur?

\begin{equation*}y=w_0x_0+...+w_nx_n + b\end{equation*}

$W$ değişkenine ağırlıklar (weights) $x$ değişkenine girdiler/öznitelikler (input/features) dersek. Yukarıda ifade ettiğimiz denklem girdiler ile çıktıların çarpılıp, sonrasında sabit bir terim ile toplanması olarak ifade edilmektedir.

Şimdi bu matris çarpımı olarak yazıdğımız denklem aslında toplam olarak da ifade edilebilir. Onu da yazalım:

\begin{equation*}y=W^Tx+b=\sum_{i=0}^n{w_ix_i}+b\end{equation*}

Bu şekil her kaynakta farklı farklı gösterilir fakat temelde yapılan şey aynıdır. Girdileri ağırlıklarla çarp ve topla,sonrasında ise sabit bir terim ekle.

Şekilde birşey daha dikkatinizi çekti mi? Şeklin en sonunda bir fonksiyon mevcut. Bu fonksiyona **aktivasyon fonksiyonu** denilmektedir. **Aktivasyon fonksiyonu** ortaya çıkan y değerinin non-lineer şekilde davranmasını sağlamaktadır. 

<div class="alert alert-warning">
Yukarıdaki şekildeki yapı genellikler <b> tek bir nöron </b>olarak tanımlanmaktadır.
</div>

Tek bir nöron denklemini aktivasyon fonksiyonunu da kullanarak tanılayalım

\begin{equation*}y'=f(W^Tx+b)=f(\sum_{i=0}^n{w_ix_i}+b)\end{equation*}

### ADIM FONKSİYONU (STEP FUNCTION)

Bu fonksiyon 1 veya 0 çıktısı verir. Genellikle ikiili sınıflandırma problemlerinde "perceptron" sınıflandırıcısı ile birlikte kullanılır. 

*Not: "Perceptron" girdileri lineer olarak birleştirerek tahmin yapmaya çalışan bir ikili sınıflandırma algoritmasıdır.*

```python
import numpy as np
import matplotlib.pyplot as plt
# Seaborn dark theme tercihimizdir :)
import seaborn as sns
sns.set_style("darkgrid")

%matplotlib notebook
```

```python
plt.figure()
plt.step([-4,0,4],[0,0,1],linewidth=3)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Adım Fonksiyonu");
```

### TANH

Çok büyük sayıları 1'e çok küçük sayıları -1'e sıkıştırır. ***"Vanishing Gradient" problemi mevcut***

```python
plt.figure()
x = np.linspace(-4,4,100) # -1 ile 1 arasında x oluşturalım
plt.plot(x,np.tanh(x),linewidth=3)
plt.xlabel("input")
plt.ylabel("output")
plt.title("tanh");
```

### ATAN

Girdileri $-\pi/2$ ile $\pi/2$ arasına sıkıştırır. *tanh* ile 0 civarında benzer bir karakteristik sergiler fakat tanh'ye göre daha yumuşaktır. ***"Vanishing Gradient" problemi mevcut***

```python
plt.figure()
x = np.linspace(-10,10,100) # -1 ile 1 arasında x oluşturalım
plt.plot(x,np.arctan(x),linewidth=3)
plt.plot(x,np.tanh(x),linewidth=3)
plt.xlabel("input")
plt.ylabel("output")
plt.title("atan")
plt.legend(["atan","tanh"]);
```

### HARD TANH

*tanh* fonksiyonuna göre işlem yükü daha azdır. 

```python
plt.figure()
x = np.linspace(-4,4,100) # -1 ile 1 arasında x oluşturalım
plt.plot(x,np.maximum(-1, np.minimum(1, x)),linewidth=3)
plt.plot(x,np.tanh(x),linewidth=3)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Hard tanh");
plt.legend(["Hard tanh","tanh"]);
```

### SIGMOID

Girdileri 0 ile 1 arasına sıkıştırır. ***"Vanishing Gradient" problemi mevcut***

```python
plt.figure()
x = np.linspace(-10,10,100) # -1 ile 1 arasında x oluşturalım
plt.plot(x,1 / (1 + np.exp(-x)),linewidth=3)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Sigmoid");
```

### RELU (RECTIFIED LINEAR UNIT)

Çok kral bir fonksiyondur :) "Vanishing gradient" problemi yoktur. Bu sebeple derin öğrenmede daha hızlı ve kesin sonuç alabilmemizi sağlar.

```python
plt.figure()
x = np.linspace(-10,10,100) # -1 ile 1 arasında x oluşturalım
plt.plot(x,np.maximum(0, x),linewidth=3)
plt.xlabel("input")
plt.ylabel("output")
plt.title("RELU");
```

### LEAKY RELU

RELU verken niye buna ihtiyaç çıktı denilebilir. Ne yazık ki bazen RELU'lar ölebiliyor (dying relu problem) ve girdilere hiç tepki vermiyor. Eğitim aşamasında bunun önüne geçebilmek için RELU'nun 0 olan kısmına hafif eğim verilir.

```python
plt.figure()
x = np.linspace(-10,1,100) # -1 ile 1 arasında x oluşturalım
plt.plot(x,np.maximum(0.02 * x, x),linewidth=3)
plt.xlabel("input")
plt.ylabel("output")
plt.title("Leaky RELU");
```
